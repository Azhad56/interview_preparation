```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
img = Image.open("strange.0.jpg")
```

```python
np_img = np.array(img)
```

```python
np_img.shape
```

```python
plt.imshow(np_img)
```

```python
model = KMeans(n_clusters=8)
```

```python
X = np_img.reshape(-1,3)
```

```python
X.shape
```

```python
model.fit(X)
```

```python
pallet =  [cluster_center for cluster_center in model.cluster_centers_.astype("uint8")]
```

```python
pallet_img = np.array(pallet)
pallet_img.shape
```

```python
plt.imshow(pallet_img.reshape(2,4,3))
```

```python
print(model.labels_.max())
print(model.labels_.min())
model.labels_
print(len(pallet))
```

```python
out_img = [pallet[label] for label in model.labels_]
```

```python
out_img = np.array(out_img)
```

```python
out_img = out_img.reshape(np_img.shape)
```

```python
plt.imshow(out_img)
```
